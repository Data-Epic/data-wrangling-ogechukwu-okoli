```python
#Loading the dataset

import pandas as pd # importig pandas
import os # importing os module this is just a safe procedure to confimr if the file is in my laptop
from datetime import datetime # importing the datetime module
file = "mock_data.csv"
# Check if file exists in the current working directory
if os.path.exists(file):
    df = pd.read_csv(file)  # reading_csv file
    totalrows = len(df)# counting the rows of the datafram
    rowstoload = int(0.75 * totalrows) # calulating 75% of the entire dataframe
    print(df.head(rowstoload))  # Load the first 75% of the rows
else:
    print(f"File {file} not found.")
```

```python
#DATA CLEANING
df = df.drop_duplicates(subset=["Customer ID","Country","Total Purchases","Sign-up Date"], keep="first") # droping duplicates if they exist
df
```

```python
df["Sign-up Date"] = pd.to_datetime(df["Sign-up Date"])# code to change the datatype of Sign-up Date columnn
df.head(5) # output for the first 5 rows
```

```python
#DATA CLEANING
# Checking  for dates in the year 2020 using pandas datetime method
replacing_date = df["Sign-up Date"].dt.year == 2020

# Replace dates in the year 2020 with NaN
df.loc[replacing_date, "Sign-up Date"]= pd.NaT
df.head(5) # printing rows
```

```python
#DATA CLEANING
# Subsetting and dropping rows where Total Purchases are less than 5 and Country is Canada
#Note ~ represents NOT in boolean expressions
df = df[~((df["Total Purchases"]<5) & (df["Country"]=="Canada"))]
df
```

```python
df = df.reset_index(drop=True) # resetting index for uniformity and the argument drop = True prevents the addition of a new index column 
df
```

```python
current_date = datetime.now()# function to check for current date
# Feature Engineering using lambda functions aka anonymous functions for subsetting data that satifies the conditions below and creating new column that stores this columns
df['Loyalty Score'] = df.apply(lambda x: 3 if x["Total Purchases"] > 20 else
                                        (2 if x["Total Purchases"] in range(10, 21) else
                                        (2 if (current_date - x["Sign-up Date"]).days > 3*365 and x["Total Purchases"]< 10 else
                                        (1 if x["Total Purchases"] < 10 else 0))), axis=1)
df.head(5) # output for the fisrt five rows in the dataframe.
```

```python
# Data aggregation and filtering
filter_dataframe = df[df["Country"] != "Mexico"]# country to filter on

# Aggregate functions on Total purchases and loyalty score calculating the mean, max, and mode while grouping on the column "Country"
grouping = filter_dataframe.groupby("Country").agg(
    avg_purchase=("Total Purchases", "mean"),
    max_purchase=("Total Purchases", "max"),
    mostcommon_loyaltyscore=("Loyalty Score", lambda x: x.mode().iloc[0])
)

grouping.head(5)
```

```python
#Data Export
# Select specified columns
specified_columns = df[["Customer ID", "Loyalty Score", "Country", "Total Purchases"]]

# Specific the date(Monaday when the task was given)
dat_e = datetime(2025, 3, 17)
actaul_date = dat_e.strftime("%d-%m-%Y")

# Saving the specified columns to a CSV file
specified_columns.to_csv(f"processed_data_{actaul_date}.csv", index=False)

# Shuffle the specified columns with a random state
specified_columns_shuffled = specified_columns.sample(frac=1, random_state=42).reset_index(drop=True)
specified_columns_shuffled.head(10)
```
