### 2. Visualize how profit is distributed across different categories using a pie chart, and explode the wedge with the highest percentage of profit across all categories.

Data: https://drive.google.com/file/d/1-cCEA9WxaLbjx8MwQJUfrQOfznnE_r5k/view?usp=sharing

```python
# To Do: Find the answer to the query using code
# reading file and importing necassary modules
import pandas as pd
import matplotlib.pyplot as plt
profit = pd.read_csv("Superstore (1).csv")
profit.info()
profit.head(10)
```

```python
profit["Category"] = profit["Category"].astype("category")# changing datype to category
profit.info()
```

```python
profit_by_category = profit.groupby('Category',observed=False)['Profit'].sum()# grouping by categroy fo find the highest category by sum
profit_by_category
```

```python
plt.figure(figsize=(10,10))
plt.pie(profit_by_category, labels=profit_by_category.index, explode=(0,0,0.2), autopct='%1.1f%%', startangle=0,shadow=True)
plt.title('Category distribution by profit')
plt.show()
```

### 3. Visualize the global sales trends of Nintendo games over time using a line plot. How have Nintendo's sales figures evolved from the 1980s to the 2000s?

Data: https://drive.google.com/file/d/1IwbdhYcqMEf4zZH0Hfgbev9pCrpvJiGk/view?usp=sharing

```python
# To Do: Find the answer to the query using code
#importing necessary modules nad reading csv file
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sales = pd.read_csv("vgsales_no_headers.csv", header = None)
renameing = {3:"Year",
        5: "Company",
        0: "ID"}
sales.rename(columns=renameing,inplace=True)
sales.head()
```

```python
sales.drop_duplicates()
```

```python
sales.describe()#summary statistics
```

```python
sales.rename(columns={10:"Value"},inplace=True)# renaming colum 10 after checking summary satistics and dicovering it has the highest in almost all rows in .describe() and ddeciding to use in constructing line graph
sales.head()
```

```python
sales['Year'] = pd.to_datetime(sales['Year'], format="%Y")# convert to datatype datetime
sales.head()
```

```python
sales["Year"] = sales["Year"].dt.strftime("%Y") # to just have the year 
sales
```

```python
sales.info()# checking dataframe info after using dt.year and noting the chabge in the datatype of the year clumn from datetime to object
```

```python

# Filtering out only Nintendo games
nintendosales = sales[sales["Company"] == "Nintendo"]

# Dropping rows with missing Year values
droping_null= nintendosales.dropna(subset=["Year"])

# Grouping by Year and sum Global Sales
trends_nintendo = droping_null.groupby("Year")["Value"].sum()

# Plotting Nintendo's global sales over time
plt.figure(figsize=(20, 5))
plt.plot(trends_nintendo.index, trends_nintendo.values, marker="o", linestyle="--", color="blue")

# Adding labels and title
plt.xlabel("Year") #labelling x axis
plt.ylabel("Global values") # labelling y axid
plt.title("Sales trend for Nintendo(1980s - 2000s)")# title
plt.grid(True)

# Showing the plot
plt.show()
```

Query 4 Answer:

The line plot indicates that the peak sales year for nintendo was in 2006 and the lowest sales was recorded in 2016

### 4. In Saudi Arabia, the 5-day workweek starts on Sunday. Analyze the impact of the weekend and weekday quote dates on the quote price.

Data: https://drive.google.com/file/d/1vROMYOn9OUrIS7ChIIzkGpjeEdxm5Pe1/view?usp=sharing

```python
# To Do: Find the answer to the query using code
#importing necessary modules
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
saudi = pd.read_csv("luxury_goods_selling_transaction - quote_sell_dataset - luxury_goods_selling_transaction - quote_sell_dataset.csv")# reading csv file
saudi.head(5)# output 5
```

```python
saudi.info()# dataframe info to know datatypes
```

```python
saudi["Quote Date"] = pd.to_datetime(saudi["Quote Date"])
saudi.info()
```

```python
saudi['Days'] = saudi['Quote Date'].apply(lambda x: x.strftime('%A'))# to convert the day in quote dates to Monday-sunday format instaed of 0-6
saudi.head(5)
```

```python
weekdays = ["Monday", "Tuesday", "Wednesday", "Thursday", "Sunday"]#list of days assigned as weekdays in saudi
weekends = ["Friday", "Saturday"] # list of days assigned as weekends
# Conditions
conditions = [
    saudi["Days"].isin(weekdays),
    saudi["Days"].isin(weekends)
]
# Choices
labels = ["Weekday", "Weekend"]
# Apply np.select
saudi['Day_Type'] = np.select(conditions, labels, default="Other")# using numpy to construct an array on some conditions i.e grouping and assigning this to a new column Day_Type
saudi.head(9)#output
```

```python
# performing statistics where Quote price ! == 0 and grouping on Day_Type and Days
group = saudi[saudi["Quote Price"]!=0.0]
grouping = group.groupby(["Day_Type","Days"]).agg(
    meanprice=("Quote Price", "mean"),
    maxprice=("Quote Price", "max"),
    minprice=("Quote Price", "min")
)
grouping# output or calling dataframe grouping
```

```python
# performing statistics where Quote price ! == 0 and grouping on Day_Type
group = saudi[saudi["Quote Price"]!=0.0]
grouping = group.groupby(["Day_Type"]).agg(
    meanprice=("Quote Price", "mean"),
    maxprice=("Quote Price", "max"),
    minprice=("Quote Price", "min")
)
grouping
```
